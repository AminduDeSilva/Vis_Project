```python
import pandas as pd

# Read and process the 2015 CSV
df_2015 = pd.read_csv("2015.csv", index_col=0).drop(["Region", "Standard Error"], axis=1)
df_2015[["Generosity", "Trust (Government Corruption)"]] = df_2015[["Trust (Government Corruption)", "Generosity"]]
df_2015.rename(columns={'Generosity': 'Trust (Government Corruption)', 'Trust (Government Corruption)': 'Generosity'}, inplace=True)
df_2015 = df_2015.loc[["Germany", "Sri Lanka"]]

# Read and process the 2016 CSV
df_2016 = pd.read_csv("2016.csv", index_col=0).drop(["Region", "Lower Confidence Interval", "Upper Confidence Interval"], axis=1)
df_2016[["Generosity", "Trust (Government Corruption)"]] = df_2016[["Trust (Government Corruption)", "Generosity"]]
df_2016.rename(columns={'Generosity': 'Trust (Government Corruption)', 'Trust (Government Corruption)': 'Generosity'}, inplace=True)
df_2016 = df_2016.loc[["Germany", "Sri Lanka"]]

# Read and process the 2017 CSV
df_2017 = pd.read_csv("2017.csv", index_col=0).drop(["Whisker.high", "Whisker.low"], axis=1)
df_2017 = df_2017.loc[["Germany", "Sri Lanka"]]

# Read and process the 2018 CSV
df_2018 = pd.read_csv("2018.csv", index_col=1).loc[["Germany", "Sri Lanka"]]
df_2018["Dystopia + residual"] = df_2018["Score"] - df_2018[df_2018.columns[2:]].sum(axis=1)

# Read and process the 2019 CSV
df_2019 = pd.read_csv("2019.csv", index_col=1).loc[["Germany", "Sri Lanka"]]
df_2019["Dystopia + residual"] = df_2019["Score"] - df_2019[df_2019.columns[2:]].sum(axis=1)

# Read and process the 2020 CSV
df_2020 = pd.read_csv("World Happiness Report 2020.csv", index_col=0)
df_2020 = df_2020[["Ladder score"] + df_2020.columns[12:].tolist()]
df_2020.insert(0, "Overall rank", df_2020["Ladder score"].rank(ascending=False))
df_2020 = df_2020.loc[["Germany", "Sri Lanka"]]

# Read and process the 2021 CSV
df_2021 = pd.read_csv("World Happiness Report 2021.csv", index_col=0)
df_2021 = df_2021[["Ladder score"] + df_2021.columns[12:].tolist()]
df_2021.insert(0, "Overall rank", df_2021["Ladder score"].rank(ascending=False))
df_2021 = df_2021.loc[["Germany", "Sri Lanka"]]

# Read and process the 2022 CSV
df_2022 = pd.read_csv("World Happiness Report 2022.csv", index_col=1)
df_2022 = df_2022[["Happiness score"] + df_2022.columns[5:].tolist() + ["Dystopia (1.83) + residual"]]
df_2022.insert(0, "Overall rank", df_2022["Happiness score"].rank(ascending=False))
df_2022 = df_2022.loc[["Germany", "Sri Lanka"]]

# Read and process the 2023 CSV
df_2023 = pd.read_csv("World Happiness Report 2023.csv", index_col=0)
df_2023 = df_2023[["Ladder score"] + df_2023.columns[11:].tolist()]
df_2023.insert(0, "Overall rank", df_2023["Ladder score"].rank(ascending=False))
df_2023 = df_2023.loc[["Germany", "Sri Lanka"]]

# Read and process the 2024 CSV
df_2024 = pd.read_csv("World Happiness Report 2024.csv", index_col=0)
df_2024 = df_2024[["Ladder score"] + df_2024.columns[3:].tolist()]
df_2024.insert(0, "Overall rank", df_2024["Ladder score"].rank(ascending=False))
df_2024 = df_2024.loc[["Germany", "Sri Lanka"]]
```

```python
df = pd.concat(
    [pd.DataFrame(df.values) for df in[df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]])

# df.index = pd.MultiIndex.from_product([["Germany", "Sri Lanka"], [str(year) for year in range(2015, 2024)]])
df.index = pd.MultiIndex.from_product([[str(year) for year in range(2015, 2025)], ["Germany", "Sri Lanka"]], names=["Year","Country"])
df.columns = ["Happiness rank", "Happiness score", "GDP per capita","Social support","Healthy life expectancy","Freedom to make life choices","Generosity","Perceptions of corruption", "Dystopia + residual"]

df["Happiness rank"] = df["Happiness rank"].astype(int)
df = df.round(3)

df.to_csv("results.csv")
df
```
